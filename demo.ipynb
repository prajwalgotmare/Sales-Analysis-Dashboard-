Plotting bar plots for Dominant operational format

```python
plt.figure(figsize = (7,5))
ax=sns.countplot(x = 'Dominant_operational_format', data = df)
plt.xlabel('Dominant Operational Format')
plt.xticks(rotation = 90)
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.25, p.get_height()+1), va='bottom', color= 'black')
```

```python
df.replace({'UnitedStatese':'UnitedStates'
             
    },inplace=True)
df['Location'] = df['Location'].str.replace(" ", "" )
df.Location.value_counts()
```

Plotting Bar Chart for Retailers as per theri location counts

```python
plt.figure(figsize=(15,6))
ax = df.Location.value_counts().plot(kind = 'bar',color ='olivedrab')
plt.xlabel('Country with maximum number of leading retailers')
for p in ax.patches:
    ax.annotate(int(p.get_height()),(p.get_x()+0.25, p.get_height()+1),
              va = 'bottom',color = 'black')
```

```python
### Filter for only Numeric Columns
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = df.select_dtypes(include=numerics)
corr_matrix = newdf.corr()


mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

#plt
f, ax = plt.subplots(figsize= (11,9))
sns.heatmap(corr_matrix, mask=mask, center=0, #add abs() to corr to find absolute values
            square=True, linewidths=.5,
            cbar_kws={"shrink": .9}, vmin=-1.2,
            vmax=1.2, cmap='coolwarm', annot=True)

# Create a heatmap of the correlation matrix
# sns.heatmap(corr_matrix, annot=True)

# Show the plot
plt.show()
```
